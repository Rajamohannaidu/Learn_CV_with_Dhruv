## 1. Image Histogram and Histogram Equalization
In this exercise, you will calculate and plot an image histogram, then perform histogram equalization.

```python
import cv2
import matplotlib.pyplot as plt

# Load a grayscale image (ensure the file path is correct)
image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

# Check if the image is loaded
if image is None:
    print("Error: Image not found. Ensure 'image.jpg' exists in the current directory.")
else:
    # Equalize histogram
    equalized = cv2.equalizeHist(image)

    # Plot original and equalized images
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(image, cmap='gray', vmin=0, vmax=255)
    plt.subplot(1, 2, 2)
    plt.title('Histogram Equalized')
    plt.imshow(equalized, cmap='gray', vmin=0, vmax=255)
    plt.show()
```

## 2. Basic Intensity Transformations
Apply and visualize log and power-law transformations.

```python

import numpy as np

# Log Transformation
c = 1
log_transformed = c * np.log(1 + image)

# Gamma Correction
gamma = 2.0
gamma_corrected = np.array(255 * (image / 255) ** gamma, dtype='uint8')

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Log Transformation')
plt.imshow(log_transformed, cmap='gray')
plt.subplot(1, 2, 2)
plt.title('Gamma Correction')
plt.imshow(gamma_corrected, cmap='gray')
plt.show()
```

## 3. Spatial Filtering
Experiment with box and Gaussian filters.

```python

# Box Filter
box_filter = cv2.blur(image, (5, 5))

# Gaussian Filter
gaussian_filter = cv2.GaussianBlur(image, (5, 5), 0)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Box Filter')
plt.imshow(box_filter, cmap='gray')
plt.subplot(1, 2, 2)
plt.title('Gaussian Filter')
plt.imshow(gaussian_filter, cmap='gray')
plt.show()
```

## 4. Color Spaces
Convert the image to different color spaces and visualize them.

```python

# Convert to HSV
hsv_image = cv2.cvtColor(cv2.imread('image.jpg'), cv2.COLOR_BGR2HSV)

# Convert to YCbCr
ycbcr_image = cv2.cvtColor(cv2.imread('image.jpg'), cv2.COLOR_BGR2YCrCb)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('HSV Image')
plt.imshow(hsv_image)
plt.subplot(1, 2, 2)
plt.title('YCbCr Image')
plt.imshow(ycbcr_image)
plt.show()
```
